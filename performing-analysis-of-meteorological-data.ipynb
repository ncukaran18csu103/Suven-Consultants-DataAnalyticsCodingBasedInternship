So very firstly we have to import important laibrary's to start with the data. Here i am using google colab.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Now upload csv file 

```python
df=pd.read_csv('/kaggle/input/weather-dataset/weatherHistory.csv')
df.head(5)
```

Import csv file into data frame as "df" and check it

```python
df.dtypes
```

```python
df['Formatted Date'].head(5)
```

As above we check the data type for all the rows in the data and we saw "Formated date" has "Objecct" data type so we have to change it into "datetime64" for further process

```python
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'],utc=True)
df['Formatted Date']
```

```python
df.describe()
```

Set "Formatted Date" as index for Analysis

```python
df = df.set_index("Formatted Date")
df
```

As we know we have "hourly data" so we have to resample it into "Monthy" and as we need only "Apparent Temperature' and "Humidity" so takeout both the column with it  

```python
data_columns = ['Apparent Temperature (C)', 'Humidity']
df_monthly_mean = df[data_columns].resample('MS').mean()
df_monthly_mean.tail()
```

So now we have Monthly data for "Apparent Temperature" and "humidity" for 10 years now its time to visualize the data. Here we can see the variation in the Apparent temperature and Humidity with time for 10 years 

```python
plt.figure(figsize=(15,6))
sns.lineplot(data=df_monthly_mean,marker="o")
plt.title("Variation in Apparent Temperature and Humidity with time")
```

**Obeservation**

From the above graph we can say that Humidity is almost constant in 10 years and temperature is also the same as peak lies on same line but little low in 2012.
